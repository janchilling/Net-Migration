```python
import numpy as np
import pandas as pd
```

```python
df_NetMigration = pd.read_csv("./Dataset/Net_Migration_Data.csv")
df_GDP = pd.read_csv("./Dataset/GDP.csv")
df_GDPPerCap = pd.read_csv("./Dataset/GDPPerCap.csv")
df_Inflation = pd.read_csv("./Dataset/Inflation.csv")
df_Unemployment = pd.read_csv("./Dataset/Unemployment.csv")
```

# Net - Migration

```python
df_NetMigration.head()
```

```python
def columnEdits(data):
    data = data.loc[data['Country Name'] == 'Sri Lanka']
    drop_columns = ['Indicator Name', 'Indicator Code', 'Country Code', 'Country Name']
    data = data.drop(drop_columns, axis = 1)
    data = data.transpose()
    return data
```

```python
df_NetMigration = columnEdits(df_NetMigration)
```

```python
# df_NetMigration.columns = ['Year']
```

```python
df_NetMigration
```

```python
df_NetMigration = df_NetMigration.reset_index()
df_NetMigration = df_NetMigration.rename(columns = {'index' : 'Year', 138 : 'Net Migration'})
df_NetMigration 
```

```python
df_NetMigration.isnull().sum()
```

```python
df_NetMigration.info()
```

```python
df_NetMigration = df_NetMigration.drop(df_NetMigration.index[:10])
df_NetMigration = df_NetMigration.reset_index(drop = True)
df_NetMigration
```

```python
df_NetMigration.to_csv("./Cleaned_Dataset/Net_Migration_Data_C.csv", index = False)
```

# GDP

```python
df_GDP.head()
```

```python
df_GDP.isnull().sum()
```

```python
df_GDP.info()
```

```python
def addRows(dataframe, fYear):
    new_data = {'Year' : list(range(1970,fYear))}
    new_rows = pd.DataFrame(new_data)
    dataframe = pd.concat([new_rows, dataframe], ignore_index = True)
    return dataframe
```

```python
df_GDP = addRows(df_GDP, 1982)
df_GDP
```

```python
df_GDP.loc[52, 'Year'] = '2022'
df_GDP
```

```python
df_GDP.to_csv("./Cleaned_Dataset/GDP_C.csv", index = False)
```
